```python
from array import array
#Creation
arr1=array("i",[1,2,3,4,5]) #signed int array
arr2=array("f",[1.2,2.3,5.4]) #flat  array
arr3=array("d",range(10)) #double array

#print(arr1,arr2,arr3)

#Access (Same as list)
arr1[0]   # first elemn
arr1[-1]  #Last element
arr[1:3]  #Slicing -->retun new array

#Modifications
arr1[0]=24
arr1.append(6)
arr1.insert(0,0)
print(arr1)

# Removal
arr1.remove(3)   #remove first occurence
print(arr1)
arr1.pop()   # remove last
print(arr1)
arr1.pop(2) # remove at index
print(arr1)

# Info
arr3.typecode   # Type of array i,f,d
arr1.itemsize   # 4 (byte per element)
arr1.buffer_info() # Memory address length

#Converstion.
arr1.tolist() # arr to Python list
arr1.tobytes() # arr to bytes object

# Extend.
arr1.extend([7,8,9])  # from iterable
print(arr1)
arr1.fromlist([10,11]) # from list
print(arr1)

#Count /Index
arr1.count(4)  #Number of occurences.
arr1.index(5) #index of first occurences


# Reverse
arr1.reverse()
print(arr1)
```

Byte array in Python : is a mutable sequence of bytes (numbers 0 to 255) used to store binary data .
Used when working with files,networking commucations,encriptions,images,low level data processing.


Byte array is a collection of bytes (8 bits integers)  ranges from 0 to 255.


```python
#b = bytearray([10,20,30])
b = bytearray([65,66,67])
print(b)
```

```python
b=bytes([65,66])
b[0]=70
```

Python has 2 binary data type:
Feature         bytes                 bytearray
Mutability      Immuatable            Mutable
Midification    Not possible          possible
Performence     Faster                slower
Usage           Fixed binary data     Editable binary data

```python
#Create Byte Arrays:
#Method 1: from List of integers.
b=bytearray([10,20,30,40])
print(b)
#Method 2:From string
b=bytearray("hello","utf-8")
print(b)
#Method 3: With sixe initialization.
b=bytearray(5)
print(b)
#Method 4: From bytes.
b=bytearray(b"python")
print(b)
```

```python
b=bytearray([10,20,30,40])
print(b)

#Accesiing Elements.
print(b[0])
print(b[-1])

#Modification : Mutable so possible
b[0]=50
print(b[0])

#Slicing allowed .
b=bytearray(b"python")
print(b[1:4])
```

```python
# Supported / common methods.
b=bytearray([65,66])
b.append(67) #Single element append
b.extend([68,69]) #Add multiple bytes.
b.insert(1,70) #Inserting values directing.
b.remove(70) # General removed based on value
b.pop() # remove last element
print(b)
```

```python
b=bytearray(b"hello")
b.clear()

print(b)
```

```python
# Convertions:
#Bytearray to String.
b =bytearray(b"hello")
s=b.decode("utf-8")
print(s)
#String to byte array
s="hello"
b=bytearray(s,"utf-8")
print(b)
#Bytearray to lst
b = bytearray(b"ABC")
#Iterating byte array
#print(list(b))
for i in b:
    print(b)
```
