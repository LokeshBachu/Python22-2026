Pandas to Read CSV files in PYTHON way:

CSV -- Comma-Seperated values files that allow storing of your tablur data.

---read_csv()  retieve data in the form of DF.

```python
import pandas as pd

df=pd.read_csv("people_data.csv")

print(df)
```

read_csv() :

is used to read data from CSV files into DF.
CSV --are plain text files where each row ---record and columns --seperated by commas(delimeters)

synstax:

pd.read_csv(filepath_or_buffer,sep=',',header='infer',index_col=None,usecols=None,engine=None,skiprows=None,nrows=None)

--Location of the CSV file ---String path (absolte or relative path)

--SEP :It stands for seperato ,default ','

--header : it accepts int,list of int,row numbers to use as columns names .

if no names are passed ---header=None then --it will display the first column as 0 and second is 1 ...

--Usecols:Retrieve only selected columns from CSV

--nrows: Number of rows to be diaplayed from dataset.

--index_col :  None --default index starts (0,1,2,3...)

--skiprows : Skip passed rows in the new DF.

```python
# read sepecific columns from csv file
df_specific_columns = pd.read_csv("people_data.csv", usecols=['First Name', 'Email'])

print(df_specific_columns)
```

```python
# Setting an index Columns 
df=pd.read_csv("people_data.csv", index_col='First Name')
print(df)
```

```python
#Handling missing values
df=pd.read_csv("people_data.csv", na_values=['Na','Unknown'])
print(df.head(2))
```

```python
df=pd.read_csv("people_data.csv",sep=",")
print(df.tail(2))
```

```python
df=pd.read_csv("people_data.csv", header=0)
print(df.head(2))
```

```python
df=pd.read_csv("people_data.csv", header=None)
print(df.head(2))
```

```python
df=pd.read_csv("people_data.csv", names=["A","B","C","D","E"])
print(df.head(2))
```

```python

#Force column data types for feature model input.
df=pd.read_csv("people_data.csv", dtype={'First Name': 'str'})
```

```python
df=pd.read_csv("people_data.csv", parse_dates=["Date of birth"])
print(df.head(2))
```

```python
# Reading Large files --Using chunks approach.
chunk_iter=pd.read_csv("employees.csv", chunksize=30)

for chunk in chunk_iter:
    print(chunk.shape)
```

```python
df=pd.read_csv("employees.csv", nrows=50)
print(df.shape)
print(df.tail(46))
```
