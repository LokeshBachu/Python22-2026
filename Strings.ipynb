Strings are ordered in nature.



```python
#Strings creation

s1="hello"
print(s1)

s2=str(345)
print(s2)

s3="A" * 5
print(s3)

s4=" "
print(s4)

s5='python'
print(s5)

s6='''python is a programming language'''
print(s6)

s7="""python is a programming language"""
print(s7)
```

```python
#Accessing characters/Elements in a string
s="python"
print(s[0])
print(s[1])
print(s[2])
print(s[3])
print(s[4])
print(s[5])
```

```python
s = "python"
print(s[0])  # Output: 'p'
#s[0] = 'P'  # This will raise a TypeError because strings are immutable
s ="P"+s[1:]  # Creating a new string by concatenating 'P' with the rest of the original string
print(s)  # Output: 'Python'
```

```python
# String slicing
s = "python"
print(s[0:3])  # Output: 'pyt'
print(s[2:5])  # Output: 'tho'  
print(s[:4])   # Output: 'pyth' (from start to index 3)
print(s[3:])   # Output: 'hon' (from
print(s[:])    # Output: 'python' (the whole string)

#Negative indexing
s = "python"
print(s[-1])  # Output: 'n' (last character)
print(s[-2])  # Output: 'o' (second last character) 
#based on negative indexing we can also do slicing
print(s[-4:-1])  # Output: 'tho' (from index -4 to -2)
print(s[-4:])    # Output: 'thon' (from index -4 to the end)
print(s[:-2])   # Output: 'pyth' (from start to index -3)
print(s[::-1]) # Output: 'nohtyp' (reversed string)
```

```python
#String operations
s1 = "Hello"
s2 = "World"
print(s1 + " " + s2)  # Output: 'Hello World' (string concatenation)
print(s1 * 3)         # Output: 'HelloHelloHello' (string repetition
print("e" in s1)  # Output: True (membership test)
```

```python
#Common string methods
s = "  Hello World  "
print(s.lower())  # Output: '  hello world  ' (convert to lowercase)
print(s.upper())  # Output: '  HELLO WORLD  ' (convert to uppercase)
print(s.title())  # Output: '  Hello World  ' (convert to title case   )
print(s.capitalize())  # Output: '  hello world  ' (capitalize first character)
print(s.swapcase())  # Output: '  hELLO wORLD  ' (swap case)
```

```python
# Check methods (return boolean)
s = "Hello World"
print(s.isalpha())  # Output: False (contains space)
print(s.isdigit())  # Output: False (contains letters and space)
print(s.islower())  # Output: False (contains uppercase letters)
print(s.isupper())  # Output: False (contains lowercase letters)
print(s.isspace())  # Output: False (contains non-space characters)
print(s.isalnum())  # Output: False (contains space )
print(s.startswith("Hello"))  # Output: True (string starts with "Hello")
print(s.endswith("World"))    # Output: True (string ends with "World")
```

```python
#Searching & Coutning
s = "hello world"
print(s.find("o"))  # Output: 4 (index of first occurrence of 'o')
print(s.find("x"))  # Output: -1 (not found)


print(s.rfind("o"))  # Output: 7 (index of last occurrence of 'o')
print(s.index("o"))  # Output: 4 (index of first occurrence of 'o')
print(s.index("x"))  # Output: ValueError (not found)

print(s.count("o"))  # Output: 2 (number of occurrences of 'o')
print(s.count("x"))  # Output: 0 (number of occurrences of 'x')
print(s.count("l"))  # Output: 3 (number of occurrences of 'l')
```

```python
#Replace & Modify
s = "hello world"
print(s.replace("world", "there"))  # Output: 'hello there' (replace 'world' with 'there')
print(s.replace("l", "L"))          # Output: 'heLLo worLd' (replace all 'l' with 'L')

# we could also use lstrip(), rstrip(), strip() methods to remove spaces
s = "  hello world  "
print(s.strip())  # Output: 'hello world' (remove leading and trailing spaces)
print(s.lstrip())  # Output: 'hello world  ' (remove leading spaces)
print(s.rstrip())  # Output: '  hello world' (remove trailing spaces)
```

```python
#Splitting & Joining
s = "hello world welcome to python"
words = s.split(" ")  # Split the string into a list of words
print(words)  # Output: ['hello', 'world', 'welcome', 'to',

# Join the list of words back into a single string
joined_string = " ".join(words)
print(joined_string)  # Output: 'hello world welcome to python'

sentences = "Hello. How are you? Welcome to Python programming."
# Split the string into sentences using different delimiters
import re
sentence_list = re.split(r'[.?!] ', sentences)
print(sentence_list)  # Output: ['Hello', 'How are you', 'Welcome to Python programming.']

#Split and join based coma as delimiter
data = "apple,banana,cherry,date"
fruits = data.split(",")  # Split the string into a list of fruits
print(fruits)  # Output: ['apple', 'banana', 'cherry', 'date']
fruits_string = "; ".join(fruits)  # Join the list of fruits with a semicolon and space
print(fruits_string)  # Output: 'apple; banana; cherry; date'
```

```python
# String formating
# Name : %s Age:    %d (Hello ,25)
#Old way of string formatting
name = "Alice"
age = 25
print("Name: %s Age: %d" % (name, age))  # Output: 'Name: Alice Age: 25'    

#.format() method
name = "Bob"
age = 30
print("Name: {} Age: {}".format(name, age))  # Output: 'Name: Bob Age: 30'

#f-strings (Python 3.6+)
name = "Charlie"
age = 35
print(f"Name: {name} Age: {age}")  # Output: 'Name: Charlie Age: 35'
```

```python
# f-strings with expressions and complex formatting
name = "David"
age = 40
print(f"Name: {name.upper()} Age: {age*2}")  # Output: 'Name: DAVID Age: 80'

name = "Eve"
age = 28
print(f"Name: {name.lower()} Age: {age/2:.1f}")  # Output: 'Name: eve Age: 14.0'
```

```python
#Formating numbers in f-strings
pi = 3.14159
print(f"Pi rounded to 2 decimal places: {pi:.2f}")  # Output: 'Pi rounded to 2 decimal places: 3.14'
print(f"Pi in scientific notation: {pi:.2e}")  # Output: 'Pi in scientific notation: 3.14e+00'
print(f"Pi with 5 total digits: {pi:5.2f}")  # Output: 'Pi with 5 total digits:  3.14' (total width of 5, including decimal point)

#Padding and width in f-strings
number = 42
print(f"Number with padding: {number:5}")  # Output: 'Number with padding:    42' (total width of 5, right-aligned)
print(f"Number with left padding: {number:<5}")  # Output: 'Number with left padding: 42   ' (total width of 5, left-aligned)

print(f"Number with zero padding: {number:05}")  # Output: 'Number with zero padding: 00042' (total width of 5, padded with zeros)

print(f"{42:5d}")  # Output: '   42' (integer with width of 5)
print(f"{42:05}")  # Output: '00042' (integer with width of 5, padded with zeros)
print(f"{3.14159:8.2f}")  # Output: '    3.14' (float with width of 8, 2 decimal places)
#Seperators in f-strings
number = 1234567890
print(f"Number with comma separator: {number:,}")  # Output: 'Number with comma separator: 1,234,567,890' 
  
#Percentages in f-strings
accuracy = 0.98765  
print(f"Accuracy: {accuracy:.2%}")  # Output: 'Accuracy: 98.77%' (converted to percentage with 2 decimal places)

#Alighnment in f-strings and layout (table formatting)
data = [
    {"name": "Alice", "age": 25, "city": "New York"},
    {"name": "Bob", "age": 30, "city": "Los Angeles"},
    {"name": "Charlie", "age": 35, "city": "Chicago"}
]

print(f"{'Name':<10} {'Age':<5} {'City':<15}")  # Header with left alignment
print("-" * 30)  # Separator line
#Use < for left alignment, > for right alignment, and ^ for center alignment
for item in data:
    print(f"{item['name']:<10} {item['age']:<5} {item['city']:<15}")    
    
```

```python
#Data and time querying and formatting
from datetime import datetime
now = datetime.now()
print(f"Current date and time: {now}")  # Output: Current date and time
print(f"Current date: {now:%Y-%m-%d}")  # Output: Current date in YYYY-MM-DD format
print(f"Current time: {now:%H:%M:%S}")  # Output: Current time in HH:MM:SS format
print(f"Current date and time: {now:%Y-%m-%d %H:%M:%S}")  # Output: Current date and time in combined format
print(f"Current year: {now:%Y}")  # Output: Current year
print(f"{now:%A, %B %d, %Y}")  # Output: Current date in 'Weekday, Month Day, Year' format
```

```python
# Comapring of 2 strings
s1 = "hello"
s2 = "world"
s3 = "hello"

print(s1 == s2)  # Output: False (comparing two different strings)

print(s1 == s3)  # Output: True (comparing two identical strings)
```

```python
# How do we join 2 strings.
s1 = "Hello"
s2 = "World"

result = s1 + " " + s2  # Concatenating s1 and s2 with a space in between
print(result)  # Output: 'Hello World'

for letter in result:
    print(letter)  # Output: Each letter of 'Hello World' printed on a new line
    
print(result[0])  # Output: 'H' (first character of the string)    
print(len(result))  # Output: 11 (length of the string 'Hello World')
```

```python
# String Membership Test
print("H" in result)  # Output: True (checking if 'H' is in 'Hello World')
print("z" in result)  # Output: False (checking if 'z' is in 'Hello World')
```

```python
text="hello"
encoded=text.encode()
print(encoded)  # Output: b'hello' (the string is encoded to bytes)
print(type(encoded))  # Output: <class 'bytes'> (the type of the encoded string is bytes)
print(encoded.decode())  # Output: 'hello' (decoding the bytes back to a string)
print(type(encoded.decode()))  # Output: <class 'str'> (the type of the decoded string is str)
```
