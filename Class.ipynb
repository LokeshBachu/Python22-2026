```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
    @classmethod
    def from_string(cls,data):
        name,age=data.split(",")
        return cls(name,int(age))    
    

        
```

```python
#print(Person.from_string("hello,28"))
p=Person.from_string("Josh,45")
print(p)
print(p.name)
print(p.age)
```
