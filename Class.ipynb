```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
    @classmethod
    def from_string(cls,data):
        name,age=data.split(",")
        return cls(name,int(age))    
    

        
```

```python
#print(Person.from_string("hello,28"))
p=Person.from_string("Josh,45")
print(p)
print(p.name)
print(p.age)
```

```python
class Dog:
    #class variables
    species='GoldenRetriever'
    
    def __init__(self,name,age):
        #Instance variables
        self.name=name
        self.age=age
        
#Create objects for Dog class
dog1=Dog("Tiger",3)
dog2=Dog("Charlie",5)

#Access class and instance variables
print(dog1.species)
print(dog1.name)
print(dog2.age)

#Modify the instance variable
dog1.name="Max"
print(dog1.name)

#Modify to class variables
Dog.species="Shibu"
print(dog1.species)
print(dog2.species)

```

Constructor:

A constructor is Python is a special method that is automatically called when a new object (instance) of a class is created.
Main purpose constructor is to initilize (set up) the objects attributes.

in python constructor is named as 

__init__()

It is also called as Dunder method(Double underscore method)

Syntax:

class ClassName:
    def __init__(self,parm):
    pass

__init__ --Special constructor method.

self --Refers to the this/current object

param --Values passed during object creation.



```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
p1=Person("Ravi",25)

print(p1.name)
print(p1.age)        
```
