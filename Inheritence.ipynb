```python
class Animal:
    def __init__(self,name):
        self.name=name
        
    def info(self):
        print("Animal name::",self.name) 
        
class Dog(Animal):
    def sound(self):
        print(self.name,"barks")
        
d=Dog("Buddy")
d.info()
d.sound()                    
```

```python
# Using super()
#To call parent class methods inside child class.
class Animal:
    def __init__(self,name):
        self.name=name
        
class Dog(Animal):
    def __init__(self,name,bread):
        super().__init__(name) #refers to the parent class
        self.bread=bread
        
d=Dog("puppy","street")

print(d.name,"  ",d.bread)        
```

```python
# Constructor inheritence.
#If child doesn't define __init__ ,parent constructor  is used.

#if child defines it's own ,parent constructor is not automatically called unless using super()
```

Method Resolution Order (MRO)
In multiple inheritence ,Python flollws C3 Linearization algorithm

Check MRO:
print(ClassName.__mro__)

print(ClassName.mro())

```python
class A: pass
class B(A): pass
class C(A): pass
class D(B,C): pass

print(D.mro())
```

```python
class A:
    def show(self):
        print("A")
class B(A):
    def show(self):
        print("B") 
        
class C(A):
    def show(self):
        print("c")
        
class D(B,C):
    pass

obj=D()
obj.show()     

print(D.mro())   

#D -->B-->c-->A -->Object
#D doesn't have show()  -->B have show()
```

Key points:
Mro determine method look up order.
import in mutiple/hybrid inheritence
python follows c3 linearlization.
left-to-right priority .
Ensure no class is searched before it's parents.

Types of Access Modifiers in Python:
Modifier type      Syntax                   Accessibility           Strictly Enforced.
Public              name                    Anywhere                No
Protected           _name                   class+subclass          No
Private             __name                  Inside class only       Partially

```python
#Public Members can be accessed anywhere --inside or outside of the class.
class Student:
    def __init__(self,name):
        self.name=name
        
    def display(self):
        print("Name::",self.name)
        
a=Student("Ravi")
print(a.name)  #Accesible
a.display() #Accesible    

#Mostly commonly used
#Defult access level in python        
```

```python
#Protected Members (_singleunderscore)
#These memebrs should not accessed outside the class or subclass
class Person:
    def __init__(self,age):
        self._age=age # Protected variable
        
class Employee(Person):
    def show(self):
        print(self._age)  #Accessible in subclass
        
p=Person(26)
print(p._age) #Still accessble   
#Python does not strictly prevent access.
#Signal for interna use.
#You expect sub class access.             
```

```python
#Private member(__double underscore)

# __variable -->_className__variable.

#It prevents accidental access fom outside class.

class BankAccount:
    def __init__(self,balance):
        self.__balance=balance #Private var
        
    def get_balance(self):
        return self.__balance
acc=BankAccount(1000)

print(acc.get_balance())     
print(acc._BankAccount__balance)
#print(_BankAccount__balance)
#print(acc.__balance) #Error      
```

Inheritence:
--IS A relation ship

A Dog is an Animal
A car is a Vehicle


Key points:
Resuse code from parent class.
Supports method overloading
Uses super() to access parent methods.
Can create tight coupling if oveused.

When to use Inheritence?
Clear hieachary and IS a relation ship

Ex
Student is Person.
Manager is an Employee.

#Composition:
Composition is when one class contains another class as an attribute.
It represents a "HAS -A Relation ship"


Ex:
A car has an Engine.
A house has a Room.

Key points:
More flexible than inheritence.
Reduce tight coupling
Objects can be replaced dynamically.

```python
class Engine:
    def start(self):
        print("Engine started")
        
class Car:
    def __init__(self):
        self.engine=Engine() # Composition
        
    def start_car(self):
        self.engine.start()
        
c =Car()
c.start_car()    


#Car does not inherit from engine.
# Car Contains Engine                    
```

Ineheritence Vs Composition.

Feature             Inheritence                     Composition.
Relationship        IS-A                            HAS-A
Coupling            Tight                           Loose
Flexibility         Less                            More.
CodeReuse           Through parentclass             Through contined objects.
Ex                  Dog is animal                   Car has Engine.

```python
class Bird:
    def fly(self):
        print("flying")
        
class Penguin(Bird): # Problem :Penguin can't fly.
    pass

# Here inheritence may not be correct.
        
```

```python
class FlyBehavaiour:
    def fly(self):
        print("flying")
        
class NoFlyBehaviour:
    def fly(self):
        print("Cannot fly")
        
class Bird:
    def __init__(self,fly_behaviour):
        self.fly_behaviour=fly_behaviour
        
    def perform_fly(self):
        self.fly_behaviour.fly()
        
penguin=Bird(NoFlyBehaviour())
penguin.perform_fly()

eagle=Bird(FlyBehavaiour())
eagle.perform_fly()      


#More flexibile
#Behaviour dynamically              
```

Use Inheritence when relationship is truely hierarchical.
Prefer Composition when you want flexibility and better design.
