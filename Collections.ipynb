Shallow copy:
Shallw copy creates a new list object but elemnts inside are still reference to the same object.

Outer list ---new
inner list -- shared

```python
#Shallow copy in lists
a =[10,20,30]
b =a.copy()
b.append(40)
print(a)
print(b)

#List variables a and b are here independent to each other.

c=[10,20,30]
d=c[:]
d.append(40)
print(c)
print(d)
#List variables c and d are here independent to each other.

e=[10,20,30]
f=list(e)
f.append(40)
print(e)
print(f)

#List variables e and f are here independent to each other.
```

```python
# Shallow copy fails when we use netsed lists

#     0      1
a =[[1,2],[3,4]]

b=a.copy()

b[0][0]=5

print(b)

print(a)

#a --->[ref1,ref2]
#b --->[ref1,ref2]

#ref1-->[1,2]
#ref1-->[3,4]

#Because inner lists share same mem eferences so shallow copy fails with nested lists
```

Deep copy:
Deep creates a complete independent copy ,include all nested objects ..

Everything is duplicated here.

Py has module called copy.

import copy

```python
import copy
a=[[1,2],[3,4]]

b=copy.deepcopy(a)
b[0][0] =5
print(a)
print(b)

#Now your original list does not change at all.
```

a ---> [ref1,ref2]
         |    |
        [1,2] [3,4]

b ---> [ref3,ref4]
         |    |
        [1,2] [3,4]  

Everything is seperate here . a ,b variables are independent here.

```python
Students=[
    ["Alice",19],
    ["Bob",22],
    ["Charlie",15]
]

print(Students)
```

```python
# Shallow copying 

backup =Students.copy()

backup[0][1]=20

print("Original:",Students)
print("Back up:",backup)

#Probelm is original and backup also changed.
```

```python
# Deep copy.
import copy

backup=copy.deepcopy(Students)

backup[1][1]=23

print("Original:",Students)
print("Back up:",backup)
```

```python
# Set operations (DSA Mathematical)
a={1,2,3,4}
b={3,4,5,6}

# --Union -all elements from both sets
union=a | b
print(union)

a.union(b)  # same as a|b
print(a.union(b))

print(a.union(b,{7,8})) # union with multiple sets
```

```python
# intersection -- common elements
intersection=a & b
print(intersection)

print(a.intersection(b))  # same as a&b
print(a.intersection(b,{4,5})) # intersection with multiple sets
```

```python
# A -B -- elements in A but not in B
difference=a-b
print(difference)

print(a.difference(b))  # same as a-b
print(a.difference(b,{5,6})) # difference with multiple sets
```

```python
# Symetric difference -- elements in A or B but not in both
sym_diff=a ^ b
print(sym_diff)
print(a.symmetric_difference(b))  # same as a^b
#print(a.symmetric_difference(b,{1,6})) # symetric difference with multiple sets
```

```python
# Inplace versions of set operations
a={1,2,3,4}
b ={3,4,5,6}

a |= b  # inplace union
a.update(b)  # same as a|=b
print(a)

a &= b  # inplace intersection
a.intersection_update(b)  # same as a&=b
print(a)

a-=b  # inplace difference
a.difference_update(b)  # same as a-=b
print(a)

a^=b  # inplace symetric difference
a.symmetric_difference_update(b)  # same as a^=b
print(a)
```
