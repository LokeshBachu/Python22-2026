```python
# Variable length Arguments
# * args (Multiple posistional arguments)

def add_numbers(*args):
    return sum(args)
    
add_numbers(1,2,3,4,5)

add_numbers(10,2)    

add_numbers()

add_numbers(1,2,3,4,5,6,7,8)

#Why use * args.

#function to flexible you can pass any no of arguments here.
```

```python
# **kwargs (Multiple Keyword Argument)
def print_info(**kwargs):
    for key ,value in kwargs.items():
        print(key,value)
        
print_info(name='uday',age=25)        

# ** allows your function to accept any no of keyword arguments.
# All keywords arguments are collected into dictionary called kwargs
```

```python
# Retrun multiple values
def calculate(a,b):
    return a+b,a-b,a/b,a*b,a%b

sum,diff,div,mul,modu=calculate(10,5)

print(sum,diff,div,mul,modu)
```

```python
# Lambda functions or Anonymous functions

#Small one line fucntions.

square = lambda x: x * x
print(square(5))
```

```python
numbers = [1,2,3,4]
result=list(map(lambda x: x * 2,numbers))
print(result)
```

```python
# Recursive functions
# calling function in same function
# Function calling itself

def factorial(n):
    if n==1:
        return 1
    return n * factorial(n-1)   # 5 -- 5 * 4---> 4 * 3 -->3 * 2-->2 *1 -->1 --> 5 * 4 * 3 * 2 * 1

factorial(5)
```

```python
# Nested functions
def outer():
    def inner():
        print("inside inner")
    inner()        
        
outer()      
```

```python
# Pass by values : immutable objects   integer or strings or tuples.

def update_number(x):
    x = x+10
    print("Inside function",x)
    
num=5
update_number(num)
print("Outside function",num)    

#Canges made inside func will not effect oiginal object 
```

```python
# Pass by refernce : Mutable objects :Lsit or dict

def add_course(courses):
    courses.append("Python")
    print("Inside function",courses)
    
course_list=["java","c++"]
add_course(course_list)
print("Outside fucntion",course_list)   

#Cahnges made inside func effect oiginal object 
```
