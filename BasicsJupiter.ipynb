```python
for i in range(5):
    print(i)
```

```python
fruits =["apple","banana","cherry"]

for fruit in fruits:
    print(fruit)
```

```python
for ch in "Hello":
    print(ch)
```

```python
students = {"name":"John", "age":20, "grade":"A"}

for key in students:
    print(f"{key}: {students[key]}")
    
for values in students.values():
    print(values)    
for key, value in students.items():
    print(f"{key}: {value}")
```

```python
#For loop with if condition
for i in range(10):
    if i % 2 == 0:
        print(f"{i} is even")
    else:
        print(f"{i} is odd")
```

```python
# Nested loop is also posible
for i in range(3):  # 0,1,2
    for j in range(2): #0,1
        print(f"i: {i}, j: {j}")
```

```python
# Break and Continue Statements
#Break statement stop the loop entirely
for i in range(10):
    if i == 5:
        break  # Exit the loop when i is 5
    print(i)
print("Loop ended.")   
#Continue statement skips the current iteration and moves to the next one
for i in range(10):
    if i % 2 == 0:
        continue  # Skip even numbers
    print(i)  # This will print only odd numbers   
 
print("Odd numbers printed.")    
#Pass: Placeholder that does nothing
for i in range(5):
    if i < 3:
        pass  # Placeholder for future code
    else:
        print(i)     
```

```python
# Else with for loop
# else block executes if loop completes normally (no break)
for i in range(5):
    #if i == 3:
        #break  # Uncommenting this will prevent else from executing
    print(i)
else:
    print("Loop completed without break.")
```

```python
#Loop with index (enumerate)
fruits = ["apple", "banana", "cherry"]
for index, fruit in enumerate(fruits):
    print(f"Index: {index}, Fruit: {fruit}")
```

```python
# Looping through two lists simultaneously (zip)
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]
for name, age in zip(names, ages):
    print(f"Name: {name}, Age: {age}")
```

```python
# List comprehensions
#Create a new list by applying an expression to each item in an existing iterable
squares = [x**2 for x in range(10)]
print(squares)  # Outputs: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

# Will use regular loops instead of comprehensions for better readability
squares = []
for x in range(10):
    squares.append(x**2)
print(squares)

even = [x for x in range(10) if x % 2 == 0]
print(even)  # Outputs: [0, 2, 4, 6
```

Stop(required)

step(optional)

Enumarate()
enemurate() allows us loop over an iterable and it's index at same time.

Inseated of manually tracking index Python does it for us.

Syntax:
enumarate(iterable,start=0)

iterable -->list ,tuple,string,range etc.
start -->number to begin counting from defult=0

```python
fruits =["apple", "banana", "cherry"]
#Regular for loop
#for fruit in fruits:
 #   print(fruit)
 
 #Using range function
for i in range(len(fruits)):
    print(fruits[i])
    
#Using enumerate  --Python is managing index internally
for index, fruit in enumerate(fruits):
    print(f"Index: {index}, Fruit: {fruit}")    
    
    
#Manual index management ---old way 
i=0
for fruit in fruits:
    print(f"Index: {i}, Fruit: {fruit}")
    i +=1 
   
#Deafult index start from 0 when can override using start parameter    
for i,fruit in enumerate(fruits, start=1):
    print(f"Index: {i}, Fruit: {fruit}")  
    
#enumarte() returns an enumerate object which is an iterator that produces pairs of index and value from the input iterable.

e=enumerate(fruits)
print(next(e))  #Outputs: (0, 'apple')  

#Strings are iterable
for index, ch in enumerate("Hello",start=2) :
    print(f"Index: {index}, Character: {ch}")       
```

```python
#index based condition without enum()
for i in range(len(fruits)):
    if i % 2 == 0:
        print(f"Index: {i}, Fruit: {fruits[i]}")
        
#index based condition with enumerate() and without range()
for i, fruit in enumerate(fruits):
    if i % 2 == 0:
        print(f"Index: {i}, Fruit: {fruit}")        
```

Enumaret is better:
avoids index errors.
clearer intent.
work with any iterable

```python
#enumerate([]) #Empty enumerate object
#enumerate(range(5)) #Working enumerate object
enumerate(fruits)

#When we don't have to use index, we can directly iterate over the elements without using enumerate or range.
```
