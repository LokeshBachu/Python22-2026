```python
#Memory management
#Variables are refrence to objects
a = [1, 2, 3]
b = a  # b references the same list as a
b.append(4)
print(a)  # Outputs: [1, 2, 3, 4]
print(b)  # Outputs: [1, 2, 3, 4]
#Check if same object
print(a is b)  # Outputs: True
#Get memory address
print(id(a))
print(id(b))
#Delete variable
del b
#print(b)  # This will raise an error since b is deleted
```

Mutable and Immutable:
Immutable (cannot change)
int,float,string,tuple,boolean,frozenset
Mutable:Can change
list,dict,set,custom objects

```python
#Immutable
x = 10
print(id(x))  # Outputs memory address of x
y =x + 5
print(id(y))  # Different memory address since integers are immutable

#Mutable
my_list = [1, 2, 3]
print(id(my_list))  # Outputs memory address of my_list
my_list.append(4)
print(id(my_list))  # Same memory address since lists are mutable
```
