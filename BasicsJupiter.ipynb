Mutable and Immutable:
Immutable (cannot change)
int,float,string,tuple,boolean,frozenset
Mutable:Can change
list,dict,set,custom objects

```python
#Immutable
x = 10
print(id(x))  # Outputs memory address of x
y =x + 5
print(id(y))  # Different memory address since integers are immutable

#Mutable
my_list = [1, 2, 3]
print(id(my_list))  # Outputs memory address of my_list
my_list.append(4)
print(id(my_list))  # Same memory address since lists are mutable
```

2.Comparision (Relational) operators

Used to compare values.They return True or Flase.

Operator                    Meaning                             Example
==                          Equal to                            5==5-->True
!=                          Not Equal                           5!=3-->True
>                           Greater than                        5>3
<                           Lessthan                            5<3
>=                         Greater than or equal                5>=3
<=                        Leass then or equal                   5<=3


```python
#Comaparision operations
a = 10
b = 20
equal = (a == b)            # Equal to
not_equal = (a != b)       # Not equal to   
less_than = (a < b)        # Less than
less_equal = (a <= b)      # Less than or equal to
greater_than = (a > b)     # Greater than
greater_equal = (a >= b)   # Greater than or equal to
print("Equal:", equal)
print("Not Equal:", not_equal)  
print("Less Than:", less_than)
print("Less Than or Equal:", less_equal)
print("Greater Than:", greater_than)
print("Greater Than or Equal:", greater_equal)
```

6.Identity operators
Check object identity (memory location)
is and is not

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

a is b  # Outputs: True(same object)
a is c  # Outputs: False (different objects)
a == c  # Outputs: True (same content)

a is not c  # Outputs: True (different objects)
b is not c  # Outputs: True (different objects)
```

```python
# small integer caching (-5 to 256)
x = 100
y = 100
x is y  # Outputs: True (same object due to caching)

a = 1000
b = 1000
a is b  # Outputs: False (different objects)
```

7.Membership operators
Check if value exists in sequence

in and not in

```python
# in operator
#Check membership in a list
fruits = ["apple", "banana", "cherry"]
"banana" in fruits  # Outputs: True
"pineapple" in fruits  # Outputs: False
"grape" not in fruits  # Outputs: True

#Sequence we found in lists dic and strings
"an" in "banana"  # Outputs: True
[1, 2] in [0, 1, 2, 3]  # Outputs: False

#Dictionary membership
data = {"name": "Alice", "age": 30}
"name" in data  # Outputs: True
"Alice" in data.values()  # Outputs: True
"Alice" not in data  # Outputs: False
```
