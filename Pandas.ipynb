```python
import pandas as pd

s=pd.Series([10,20,30,40],index=['a','b','c','d'])
print(s)

#Index -->a,b,c,d
#Values -->10,20,30,40
```

```python
#Accessing Using Position  --(Integer Index)  -- iloc()   --elements based on position.
# Try to access labeled based Name Error.Out of Index --Index out of exception#
#print(s.iloc[5])
print(s.iloc[1])

#Postion starts from 0 

#Mutiple positions using iloc()
print(s.iloc[[0,2]])

#Accessing as A Range (Slicing)
print(s.iloc[1:3])

#Start index is included here and End index is excluded.
```

```python
# Using Labels   -- loc()
# Index which is not there or use integer based --Key error#

print(s.loc['b'])
print(s.loc[['b','a']])
print(s.loc['b':'d'])

#print(s.loc['e'])
```

```python
import pandas as pd

df=pd.DataFrame({
    'Name':['Alice','Bob','Charlie','Justin'],
    'Age':[22,25,30,55],
    'City':['Newyork','Bangalore','Hyd','London'],
})

#Setting a Name as index.
df.set_index('Name', inplace=True)
print(df)

print(df.index)

# Resetting index to default
df.reset_index(inplace=True)
print(df)
print(df.index)
```

```python
# Rename Index
df.index.name='RowID'
df.rename(index={0:'row1',1:'row2',2:'row3',3:'row4'}, inplace=True)
print(df)

# Soting index.
df.sort_index(inplace=True)

# Droping index.
df.drop(index='row2', inplace=True)
print(df)
```

```python
#Index Alignment   ---Pandas automatically aligns data based on the index labels when performing operations between Series or DataFrames. This means that when you add two Series together, pandas will match the index labels and perform the operation only on those labels that are present in both Series. If an index label is missing in one of the Series, the result for that label will be NaN (Not a Number). This feature allows for flexible data manipulation and ensures that operations are performed correctly even when the data is not perfectly aligned.
df1=pd.DataFrame({
    'A':[1,2,3]},index=[0,1,2])

df2=pd.DataFrame({
    'A':[4,5,6]},index=[1,2,3])

print(df1)
print(df2)
print(df1+df2)

#Indexes are immutable in pandas, which means you cannot change the index labels directly. However, you can create a new index or modify the existing one by using methods like set_index(), reset_index(), or rename(). If you try to assign a new value to an index label directly, it will raise an error. For example, if you have a DataFrame df and you try to do df.index[0] = 'new_label', it will raise a TypeError because the index is immutable. Instead, you should use df.rename() or df.set_index() to change the index labels.
#df1.index[0]=2
```
