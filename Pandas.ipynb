Basic Differences:

Feature               .loc                       .iloc
Type                  Labeled-based indexing     Postion based indeing (Integer numbers)

Slice                 Inclusive                  Excusive

```python
import pandas as pd
df=pd.DataFrame({
    'Name':['Alice','Bob','Charlie','Justin'],
    'Age':[22,25,30,55],
    'City':['Newyork','Bangalore','Hyd','London'],
},index=["a","b","c","d"])

#print(df)
#print(df.index)


#label-based indexing .loc()  --df.loc[row_label,column_label]
# Single row
print(df.loc["c"])
#Accessing multiple rows
print(df.loc[["a","c"]])

#Access specific value using loc
print(df.loc["c","City"])

# Access Multiple columns
print(df.loc[["a","c"],["Name","City"]])
print(df.loc[:,["Name","City"]])

# Row + column access
print(df.loc["a":"c",["Name","City"]])   # Slicing --inclusive way of slicing "a":"c"  -->c is included here. and b is taged along\
    
#Boolean filtering using loc
print(df.loc[df['Age'] > 25])    

# Conditionally select our columns
df.loc[df['Age'] > 30, ['Name','City']]

# Assign Values -- Just adding a new value to a specific cell.
df.loc["a","Age"]=20

print(df)

# Add entire row it self
df.loc["e"]=["David",40,"Chicago"]
print(df)
```

```python
# iloc -- Position based indexing  --Use integer based indexing to select data by position.(based on 0-based index)
# Basic synatx : df.iloc[row_index,column_index]
#print(df)
# Single row
print(df.iloc[0])
# Accessing multiple rows
print(df.iloc[[0,2]])

#Specific value using iloc
print(df.iloc[2,2])

# Rows + column selection
df.iloc[0:3,0:2]  # Slicing -- exclusive way of slicing 0:3 --> 3 is excluded here. and 0,1 are taged along.

#Assign values using iloc
df.iloc[0,1]=21
print(df)
```

```python
# Labels VS Postions ---loc vs iloc

#df.loc["a"]  # Label based access
df.iloc[0]   # Position based access

# Slice behaviour 
df.loc["a":"c"]  # Label based slicing --inclusive way of slicing
df.iloc[0:3]   # Position based slicing --exclusive way of slicing

# Integer Index confusions 
df1=pd.DataFrame({"A":[10,20,30]},index=[10,20,30])
df1.loc[10]  # Label based access  ---label works fine
#df1.iloc[10]  # Position based access  ---Position ---Error out of index

df1.iloc[2]

# Error 
#df1.loc[100]  # Label based access --Key error
#df1.loc["k"]

#df.iloc[10] # InterError
```

```python
# Real usage of loc and iloc
df=pd.DataFrame({
    'Name':['Alice','Bob','Charlie','Justin'],
    'Age':[22,25,30,55],
    'City':['Newyork','Bangalore','Hyd','London'],
},index=["a","b","c","d"])
print(df)
# Mixed conditions using loc
df.loc[(df['Age'] > 25) & (df['City'] == 'Hyd')]
# Select Rows using List and also slice
df.loc[["a","c"],"Name":"City"]  # Mixed selection using list and slice. Here we are selecting rows using list and columns using slice.

# Reverse
df.loc[::-1]

# Select 4 th row  df.iloc[3]  ---Select nth row df.iloc[::n-1]
df.iloc[::2]
```

```python
# Boolean with .iloc --may not work
#df.iloc[df["Age"] > 25]   ----Not Allowed --Value Error

df.iloc[df["Age"].values > 25]   # Allowed --Works fine. It is a boolean indexing with iloc.
```

```python
# Filter datafame using multiple conditions
import pandas as pd
df=pd.DataFrame({
    'Name':['Alice','Bob','Charlie','Justin'],
    'Age':[25,30,35,28],
    'City':['Newyork','Bangalore','Hyd','London'],
    'Salary':[50000,60000,70000,80000]
})

# Age > 30 and City = 'Hyd  --&& --and operator
print(df[(df["Age"]>30) & (df["City"] == "Hyd")])

print(df[(df["City"] == 'Bangalore') | (df["Salary"] > 60000)])  # OR operator

print(df[df['City'].isin(['Hyd','London'])])  # Using isin() method to filter data based on a list of values.   


print(df[(df["Age"] > 25) & (df["City"].isin(['Hyd','London'])) & (df['Salary'] > 60000)])  # Multiple conditions with isin() method

```

```python
import pandas as pd
data1 = {'Key1':['K0','K1','K2','K3'],
         'Name':['Alice','Bob','Charlie','David'],
         'Age':[25,30,35,40]}
data2={'Key1':['K0','K1','K2','K3'],
         'City':['Newyork','Bangalore','Hyd','London'],
         'Qualification':['Graduate','Postgraduate','PhD','Graduate']}

df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)

print(df,"\n\n",df1)
```

```python
# Merging of DF using a key.
result=pd.merge(df,df1,on='Key1')
print(result)
```

```python
import pandas as pd
data1={'Key':['K0','K1','K2','K3'],
       'Key1':['K0','K1','K0','K1'],
       'Name':['Alice','Bob','Charlie','David'],
       'Age':[25,30,35,40]}

data2={'Key':['K0','K1','K2','K3'],
       'Key1':['K0','K0','K0','K0'],
       'Address':['123 Main St','456 Oak Ave','789 Pine Rd','321 Elm Dr'],
       'Qualification':['Graduate','Postgraduate','PhD','Graduate']}

df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)

print(df,"\n\n",df1)
```

```python
res=pd.merge(df,df1,how='left',on=['Key','Key1'])
print(res)
```

```python
res=pd.merge(df,df1,how='right',on=['Key','Key1'])
print(res)
```

```python
res=pd.merge(df,df1,how='outer',on=['Key','Key1'])
print(res)
```

```python
res=pd.merge(df,df1,how='inner',on=['Key','Key1'])
print(res)
```

sort_values()  ---makes us to easy to sort our df by a single column--sorts in ascending order --We can customize this to desc order.

```python
data = {'Name':['Alice','Bob','Charlie','john'],
       'Age':[25,30,35,45],
       'Score':[85,90,95,80]}

df=pd.DataFrame(data)

sorted_df=df.sort_values(by='Age',ascending=False)
print(sorted_df)
```

By ---column to use fo sorting.
ascending : boolean --True(default) asc and flase --desc.
inplace: If true the original  df is modified otherwise a new sorted df is returned.
na_poisition: Control where NAN values to be placed.
igonor_index: if True : resets the index after sorting
