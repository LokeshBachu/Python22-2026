Opeartor Overloading in Python:
Allow us to define how operators(+,-,*,/,==,<) behave for custom objects using magic methods (dunder methods)

Python internally convert operators into method calls.

a+b  --> a.__add__(b)

Why:
make user defined objects behave like build in type.

```python
class Vector:
    def __init__(self,x,y):
        self.x=x
        self.y=y
        
    def __add__(self,other):
        return Vector(self.x+other.x,self.y+other.y)
    
    def __repr__(self):
        return f"Vector({self.x},{self.y})"
    
v1=Vector(2,3)
v2=Vector(4,5)

print(v1+v2)

            
```

Operator            MagicMethod

+                   __add__
-                   __sub__
*                   __mul__
/                   __truediv__
//                  __floordiv__
%                   __mod__
**                  __pow__
==                  __eq__
!=                  __ne__
[]                  __getitem__
in                  __contain__

```python
# Overloading Comparision operator.
class Student:
    def __init__(self,marks):
        self.marks=marks
        
    def __gt__(self,other):
        return self.marks>other.marks
    
s1=Student(95) 
s2=Student(80)

print(s1>s2)
           
```

```python
#Overloading ==
class Person:
    def __init__(self,name):
        self.name=name
        
    def __eq__(self,other):
        return self.name==other.name
    
s1=Person("Vijay") 
s2=Person("Ravi")

print(s1==s2)
```

```python
#Overloading [] (Indexing)
class Mylist:
    def __init__(self,data):
        self.data=data
        
    def __getitem__(self,index):
        return self.data[index]
    
obj=Mylist([10,20,30])

print(obj[1])
```

Detect Mixed Data types and Fix them in Pandas.

Mixed Data types in column (Ex number+strings+missing values) are common in real world datasets.

--Calculation errors.
--Sorting/Filtering.
--Performce Problems
--Unexcepted behaviour.


Resolve thease problems by using the concept called how to detect and fix your mixed data types.

What are mixed Data types?

```python
#What are mixed Data types?
import pandas as pd
df=pd.DataFrame({
    "age":[25,"30",35,"unknown",None]
})
print(df)
print(df.dtypes)

#Even though ages are numeric ,coulmn type object beacuse of mixed values.
```

```python
# How to detect Mixed Data types.
#1.Check dtypes.
print(df.dtypes)

#If numeric column shows as object .It might have mixed datatypes in it.
```

```python
#Method:2 Check Unique Python Types.
print(df["age"].apply(type).value_counts())

#Very much reliable detection technic.
```

```python
#Method 3 :Use infer_objects()
df=df.infer_objects()
print(df.dtypes)
print(df)

#Attempts to convert object columns to better types automatically.
```

```python
#Method4 Check with pd.to_numeric(errors='coerce')

pd.to_numeric(df["age"],errors="coerce")
```

```python
#Fixing Mixed Data types.
#Case 1:nuneric Column Stored as object
df["age"]=pd.to_numeric(df["age"],errors="coerce")
print(df)

#Converts valid numbers.
#Converts invalid values to Nan
```

```python
# Case 2 :Remove Invalid Vlaues.

df=df[pd.to_numeric(df["age"],errors="coerce").notna()]
print(df)
```

```python
#Case 3:Replace Invlaid Strings First.
df["age"]=df["age"].replace("unknown",None)
df["age"]=pd.to_numeric(df["age"])
print(df)
```

```python
#Case 4: Clean before Convertion (Applies to common real data)
df=pd.DataFrame({
    "salary":["1,00,000","200000","55,500"]
})
df["salary"]=df["salary"].str.replace(",","")
df["salary"]=pd.to_numeric(df["salary"])
print(df)
```

```python
#Case 5: Date column Mixed Types.
df["date"]=pd.to_numeric(df["date"],errors="coerce")
```

```python
#Case 6:Boolean Column Mixed types.
df["active"]=df["active"].map({
    "yes":True,
    "No":False
})
```

```python
#Detect Mixed tpes to Entire DF.
for col in df.columns:
    print(col)
    print(df[col].apply(type).value_counts())
    print("-" * 20)
    
# This will help us to identity problamatic columns in DF    
```

```python
df=pd.DataFrame({
    "price": ["100","200","N/A","300"]
})

print(df)
print(df.dtypes)

#Convert safely
df["price"]=pd.to_numeric(df["price"],errors="coerce")

#Fill missing values.
df["price"]=df["price"].fillna(df["price"].median())

print(df)
print(df.dtypes)
```

```python
import pandas as pd

data_frame=pd.DataFrame([['vinny',10],['nick',15],['juli',14.8]],columns=['Name','Age'])

#Traverse data frame to detect mixed datatypes

for column in data_frame.columns:
    
    print(column,':',pd.api.types.infer_dtype(data_frame[column]))
    
# Transform mixed data types to single data type
data_frame[column]=data_frame[column].astype(int)

#Traverse data frame to detect data types after fix.
for column in data_frame.columns:
    
    print(column,':',pd.api.types.infer_dtype(data_frame[column]))    
```

```python
import pandas as pd

data_frame=pd.DataFrame([['vinny',10],['nick',15],['juli',14.8]],columns=['Name','Age'])

#Traverse data frame to detect mixed datatypes

for column in data_frame.columns:
    
    print(column,':',pd.api.types.infer_dtype(data_frame[column]))
    
    
# Transform mised data types to single data type
data_frame[column]=data_frame[column].apply(lambda x: pd.to_numeric(x,errors='ignore'))

#Traverse data frame to detect data types after fix.
for column in data_frame.columns:
    
    print(column,':',pd.api.types.infer_dtype(data_frame[column]))     
```

a.__add__(b)

if that returns Notimplemented . 

b.__radd__(a)

both fail --->TypeError.

Proper type handling code is very much important.

--Check type of other
--Return NotImplemented if unsupported.


```python
class Vector:
    def __init__(self,x):
        self.x=x
        
    def __add__(self,other):
        if isinstance(other,Vector):
            return Vector(self.x+other.x)
        return NotImplemented    
```

```python
class Number:
    def __init__(self,value):
        self.value=value
        
    def __add__(self,other):
        if isinstance(other,Number):
             return Number(self.value+other.value)
        return NotImplemented   
    def __radd__(self,other):
        return Number(self.value+other)     
```

```python
n=Number(10)
print(5+n)
```

```python
class Money:
    def __init__(self,amount):
        self.amount=amount
        
    def __add__(self,other):
        if isinstance(other,Money):
            return Money(self.amount+other.amount)
        return NotImplemented
    def __str__(self):
        return f"${self.amount}"        
```

```python
m1=Money(100)
m2=Money(50)
print(m1+m2)
```
